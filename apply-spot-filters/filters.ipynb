```python
from pycoingecko import CoinGeckoAPI
from datetime import datetime

cg = CoinGeckoAPI()

def get_historical_mcap(coin_id, date_str):
    """
    Fetch historical market cap in USD for coin_id on date_str (format 'dd-mm-yyyy').
    Returns None if data not available.
    """
    try:
        hist_data = cg.get_coin_market_chart_range_by_id(
            id=coin_id,
            vs_currency='usd',
            from_timestamp=int(datetime.strptime(date_str, "%d-%m-%Y").timestamp()),
            to_timestamp=int(datetime.strptime(date_str, "%d-%m-%Y").timestamp()) + 86400
        )
        # hist_data structure: {'market_caps': [[timestamp, mcap], ...], ...}
        if hist_data and "market_caps" in hist_data and hist_data["market_caps"]:
            # Return the first available market cap value for that day
            return hist_data["market_caps"][0][1]
    except Exception as e:
        print(f"Error fetching mcap for {coin_id} on {date_str}: {e}")
    return None
```

```python
coins_list = cg.get_coins_list()
symbol_to_id = {coin['symbol'].upper(): coin['id'] for coin in coins_list}
```

```python
import time

date_str = "10-08-2024"
token_mcap_on_date = {}

for token in token_data.keys():
    coin_id = symbol_to_id.get(token.split('_')[0])  # assuming token like "BTC_USDT"
    if coin_id:
        mcap = get_historical_mcap(coin_id, date_str)
        token_mcap_on_date[token] = mcap
        print(f"{token}: Market Cap on {date_str} = {mcap}")
        time.sleep(1.2)  # rate limit to avoid blocking
    else:
        print(f"{token}: No coin ID found on CoinGecko")
```
